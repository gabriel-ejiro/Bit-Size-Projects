### Months in Correct Order


```python
import pandas as pd
```

```python
# data from https://www.kaggle.com/datasets/mchirico/montcoalert
df = pd.read_csv("C:\Users\ejiro\Downloads\911.csv")
```

```python
df.head()
```

```python
(df
 .sample(1000)
 .assign(month_name=lambda df: pd.DatetimeIndex(df['timeStamp']).month_name())
 .pipe(lambda df: pd.crosstab(index=df['title'], columns=df['month_name']))
)
```

```python
(df
 .sample(1000)
 .assign(month_name=lambda df: pd.DatetimeIndex(df['timeStamp']).month_name())
 .astype({'month_name': 'category'})  # Convert to categorical data
 .assign(month_name=lambda df: df['month_name'].cat.reorder_categories(['January','February','March','April','May','June','July','August','September','October','November','December']))
 .pipe(lambda df: pd.crosstab(index=df['title'], columns=df['month_name']))
)
```
